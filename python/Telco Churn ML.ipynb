## Improving the model
* There is a class imbalance in the data. The model is biased to non-churners, as the dataset has much more non-churners than churners.
* We have to make the model pay extra attention to churn customers as they are fewer.

```python
from sklearn.linear_model import LogisticRegression 
model = LogisticRegression(max_iter = 1000, class_weight = 'balanced') 
# class_weight = 'balanced' tells the model to automatically adjust the weights of the classes (churn vs no churn) to handle any imbalance in the target variable, which can help improve performance when one class is much more frequent than the other.

model.fit(X_train_scaled, y_train)
```

# Why is the model predicitng churn?

```python
coefficients = pd.DataFrame({
    'Feature': X_encoded.columns,
    'Coefficient': model.coef_[0]
})
# this creates a df called coefficients with two columns: 'Feature' and 'Coefficient'
# model.coef_ is an array of shape (1, n_features) containing the coefficients for each feature in the logistic regression model. We take the first row (model.coef_[0]) to get the coefficients for our features and create a DataFrame to easily analyze and sort them.
# A positive coefficient means churn likelihood increases as the feature value increases, while a negative coefficient means churn likelihood decreases as the feature value increases.
coefficients = coefficients.sort_values(
    by='Coefficient',
    ascending=False
)
# this sorts the coefficients DataFrame in descending order based on the 'Coefficient' column, so that the features with the highest positive impact on churn are at the top and those with the most negative impact are at the bottom.
coefficients.head(15)
```

* Some of these make sense like fiber optic customers (pay more, expect more), paperless billing (more digitally engaged, easier to switch providers), electronic check (manually make electronic payments instead of automated billing), and monthly charges (higher monthly charge = higher churn possibility).
* Some dont make sense, like having a partner. The model isnt saying having a partner causes churn, but that in THIS dataset, partnered customers correlate slightly with churn. Also, parterned customers maybe using fiber internet more, have higher monthly charges, etc.

# Model Tuning 
* Using probabilities instead of predictions (you will get probability of churn ex 0.82 instead of yes/no)

```python
# model.predict_proba(X_test_scaled) predicts how likely a customer is to churn
# [:, 1] returns only the probabilities of churn (the second column of the output, where the first column would be the probability of not churning)
# All results are stored in the variable y_probs
y_probs = model.predict_proba(X_test_scaled)[:, 1]
```

```python
import numpy as np

# Logistic Regression uses 0.5 as default threshold (> 0.5 churn, <= 0.5 no churn) so we changed the threshold to 0.35 to see if it improves recall (catching more churn customers)
# np.where is similar to an if-statement: if p > 0.35, then predict 1 (churn), else predict 0 (no churn)
y_pred_new = np.where(y_probs > 0.35, 1, 0)
```

```python
print(classification_report(y_test, y_pred_new))
print(confusion_matrix(y_test, y_pred_new))
```

* Recall decreased but precision increased. 
* This results in catching almost every churn customer, but the trade-off is now there are many false alarms
